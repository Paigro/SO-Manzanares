Realiza dos ejecuciones con el comando strace para identificar cómo se reserva la memoria del array 
ptr. Consulta el mapa de memoria del proceso en /proc para completar la siguiente tabla para la

manual de malloc(3). 
- Caso A. strace ./ejercicio7 1 
- Caso B. strace ./ejercicio7 1024 

```python
#include <stdio.h> 
#include <unistd.h> 
#include <stdlib.h> 
 
int main(int argc, char *argv[]) { 
    int factor    = atoi(argv[1]); 
    int elementos = factor * 1024; 
    printf("Inicializando región de memoria\n"); 
    int * ptr = (int *) malloc(elementos * sizeof(int)); 
    for (size_t i = 0; i < elementos; i++){ 
        ptr[i] = 1; 
    } 
    sleep(600); 
    return 0; 
} 
```
